```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
import glob

# pd.read_csv("./logs_csv/DQN/2x2_10x")

path1 = "./logs_csv/DQN/2x2_10x"
path2 = "./logs_csv/DQN/3x3_10x"
path3 = "./logs_csv/DQN/4x4_10x"

df1 = pd.concat(map(pd.read_csv, glob.glob(path1 + "/*.csv")), axis=0)
df2 = pd.concat(map(pd.read_csv, glob.glob(path2 + "/*.csv")), axis=0)
df3 = pd.concat(map(pd.read_csv, glob.glob(path3 + "/*.csv")), axis=0)

df1 = df1.drop(columns=["Wall time"])
df2 = df2.drop(columns=["Wall time"])
df3 = df3.drop(columns=["Wall time"])

# df3["Step"] = df2["Step"].values

df3
```
